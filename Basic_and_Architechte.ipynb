Spark -

Spark is a Distributed Processing program/system which distribute our task into several machines and perform the task efficiently. In layman's term Spark is a Data Processing System, through whcih we can process huge volume of dataset.

Overview of Apache Spark -

Apache Spark is a lightning-fast unified analytics engine for big data and machine learning. It was originally developed at UC Berkeley in 2009.

Lets understand this by using a Simple restuarant kitchen example:

-> When any order received from the customer for any dish,The main Chef will split the task to complete the menu quckily and serve into the table.
    So what basically he did, he assigned the task to the junior Chef in the kitchen like someone witll cut the veggies and someone will prepare the meat. In this case everyone is working efficiently and simulteneously. All these work is going to be managed by the restaurant manager and he is responsible for provide the availability of the Jr. Chef's. He will decide howmany workers needed to perform the task.

->  Like this above process Spark also runs the system efficiently.

-> There is a specific Driver program which will split the task and assign it to Worker Node, and everything is going to be managed by the Cluster manager who is responible for work to be done efficiently.

    

Driver Program : Is responsible for analyzing the data and splitting into different Worker Nodes.

Worker Node : Exucute the task which are assigned by the Driver Program.

Cluster Manager : Responsible for avaialbility of the Worker Node and monitoring the works.

Spark Context : Process through whcih Driver Program and Worker Node communicate with each other.


Architechre Flow : 

-> When we load any dataset like CSV file, it don't store the file, it just read the file and preparea LOGICAL PLAN as for of DAG(Directed Acyclic Graphs).

-> All these DAG stages are further a combination of RDD (Resilient Distributed Dataset).

-> -RDD is immutable in nature, that means once create it can not be modified.
   - Each RDD split into multiple partitions which are running in different NODES of Cluster.
   - RDD has two operation TRNSFORMATION and ACTION.
   - Any TRNSFORMATION to RDD is Lazy that means TRNSFORMATION is only going to happen when you called the ACTION.
   - RDD is the fundamental Data structre of Apache Spark.


Apache Spark Pro's :

-In Memeory computation:
    Like Hadoop it doesn't run the process in Disk memory, thats the reason we replaced spark over hadoop. Spark run the process in Memory.

-Fault tolerance: 
    It create multiple RDD over RDD so we can traceback whenever we want.

-Paritioning:
    It distribute the data into Cluster of machine to perform any task.

-Lazy Evaluation: 
    Datasets are readed stored by multiple Transformations and upon that a Logical plan is create. Unit unless system doesn't hit by the ACTION it is not going to process the transformation.
    means Transformation is going to stored by the logical plan, it is only going to execute when we do the ACTION.
